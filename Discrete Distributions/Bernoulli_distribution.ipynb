```python
# Load Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters for the Bernoulli distribution
p = 0.5  # Probability of success
n = 10000  # Number of trials
```

```python
# Simulate Bernoulli trials
data = np.random.binomial(1, p, n)
```

```python
# Display the results
print(f"Simulated Bernoulli trials (first 20 values):\n{data[:20]}")
```

```python
# Calculate frequencies
unique, counts = np.unique(data, return_counts=True)

# Plot the results
plt.bar(unique, counts, color=['blue', 'orange'], tick_label=['0 (Failure)', '1 (Success)'], alpha=0.7, edgecolor='black')
plt.title(f"Bernoulli Distribution Simulation\n(p = {p}, n = {n})", fontsize=16)
plt.xlabel("Outcome", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
