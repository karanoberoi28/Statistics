```python
# Load Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Average rate of customers (λ)
lambda_rate = 3  # Average customers per hour

# Number of simulations (hours)
num_hours = 100
```

```python
# Simulate Poisson-distributed customer counts
customer_counts = np.random.poisson(lambda_rate, num_hours)
```

```python
# Display the results
print(f"Simulated customer counts for {num_hours} hours:\n", customer_counts)
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.hist(customer_counts, bins=range(0, max(customer_counts) + 2), alpha=0.7, color='blue', edgecolor='black', align='left')
plt.title(f"Poisson Distribution Simulation\n(λ = {lambda_rate}, {num_hours} Hours)", fontsize=16)
plt.xlabel("Number of Customers in an Hour", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
